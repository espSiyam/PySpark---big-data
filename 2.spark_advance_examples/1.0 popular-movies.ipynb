```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("PopularMovies")
sc = SparkContext(conf = conf)
```

```python
lines = sc.textFile("/home/siyam/Desktop/github/0.1_personal_projects/spark_course/data/u.data")
movies = lines.map(lambda x:(int(x.split()[1]), 1))
movies_count = movies.reduceByKey(lambda x,y: x+y)

flipped = movies_count.map(lambda key_value: (key_value[1], key_value[0]))
sorted_movies = flipped.sortByKey(ascending=False)
key_value_sorted = sorted_movies.map(lambda value_key: (value_key[1], value_key[0]))

results = key_value_sorted.collect()

for result in results:
    print(result)
```
